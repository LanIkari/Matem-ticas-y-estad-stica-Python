```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
rnd=np.random.default_rng(2099)
datos=np.arange(15)
print(datos)
probabilidades=np.array([0.3, 0.2, 0.1, 0.05, .05, 0.05, 0.05, 0.025,
0.025, 0.025, 0.025, 0.025, 0.025, 0.025, 0.025])
assert round(sum(probabilidades), 10) == 1.0,"la probabilidad debe de ser 1" #Prueba unitaria
seleccion=rnd.choice(datos,p=probabilidades,replace=True)
print(seleccion)

seleccion_datos=rnd.choice(datos,p=probabilidades,replace=True,size=(4,3))
print(seleccion_datos)
```

```python
rndDos=np.random.default_rng(2000)
rndFlotantes=rndDos.random(size=(3,3))
print(rndFlotantes)

rndF=np.random.uniform(1.2,1.60,size=10)
print(rndF)
rndEnteros=rndDos.integers(5,15,size=8)
print(rndEnteros)

distribucion=rndDos.random(size=100)
fig,ax=plt.subplots()
ax.hist(distribucion)
ax.set_title("Histograma de numeros aleatorios")
ax.set_xlabel("valor")
ax.set_ylabel("Frecuencia")
plt.show()
print(distribucion)
```

```python
rnd=np.random.default_rng(2033)
mu=5.0 #Valor medio
sigma=3#Desviacion estandar

rnds=rnd.normal(loc=mu,scale=sigma,size=100)
fig,ax=plt.subplots()
ax.hist(rnds,bins=8)
plt.show()
```
