Metodo NEWTON RAPSODI

```python
def newton_iters(x0,num_iteraciones):
    iteraciones=[x0]
    #Calcular el margen de error de la ecuacion es
    errores=[abs(x0-1.)]
    for i in range(num_iteraciones):
        x0=x0-(x0*x0-1.)/(2*x0)
        iteraciones.append(x0)
        errores.append(abs(x0-1))
    return iteraciones,errores
```

```python
it,er=newton_iters(2,5)
fig,axs=plt.subplots(1,2,tight_layout=True)
axs[0].plot(it,"x-")
axs[0].set_title("Iteraciones")
axs[0].set_label("i")
axs[0].set_ylabel("x_i")

axs[1].plot(er,"o-")
axs[1].set_title("Errores")
axs[1].set_label("i")
axs[1].set_ylabel("x_i")
```

Graficas en 3D

```python
X= np.linspace(-2, 2)
Y = np.linspace(-1, 1)
x, y =np.meshgrid(X, Y)
z= x**2*y**3
fig = plt.figure()
ax = fig.add_subplot(projection="3d")
ax.plot_surface(x, y, z)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.set_title("Graph of the function f(x) = x^2y^3")
```
